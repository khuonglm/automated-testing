```python
import matplotlib.pyplot as plt
import numpy as np
import json
import os
```

```python
# Sample data - replace with your actual data
path = '/Users/khuongle/Documents/automated-testing/test_cov/our_cov/stats_summary_2025-06-12_22-09-02_{i}.json'
steps = [0, 100, 200, 300, 400, 500, 600, 700, 800, 900, 999]
line_coverage = []
branch_coverage = []
for i in steps:
    new_path = path.format(i=i)
    with open(new_path, 'r') as f:
        data = json.load(f)
    line_coverage.append(data['stats']['line_coverage'])
    branch_coverage.append(data['stats']['branch_coverage'])

# Create subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# First subplot: Iterations vs Line Coverage
ax1.plot(steps, line_coverage, marker='o', linewidth=2, markersize=6, color='green')
ax1.set_xlabel('Number of Iterations', fontsize=12)
ax1.set_ylabel('Line Coverage (%)', fontsize=12)
ax1.set_title('Line Coverage vs Iterations', fontsize=14, fontweight='bold')
ax1.grid(True, alpha=0.3)

# Second subplot: Iterations vs Branch Coverage
ax2.plot(steps, branch_coverage, marker='s', linewidth=2, markersize=6, color='red')
ax2.set_xlabel('Number of Iterations', fontsize=12)
ax2.set_ylabel('Branch Coverage (%)', fontsize=12)
ax2.set_title('Branch Coverage vs Iterations', fontsize=14, fontweight='bold')
ax2.grid(True, alpha=0.3)

# Set axis limits
# ax1.set_xlim(0, 1000)
# ax1.set_ylim(0, 1)
# ax2.set_xlim(0, 1000)
# ax2.set_ylim(0, 1)

# Add some styling
plt.tight_layout()

# Display the graph
plt.show()
```
